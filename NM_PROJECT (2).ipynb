#REGRESSION MODEL USING MACHINE LEARNING

```python
data={
    'country':['USA','India',"UK","China","Brazil","Europe","Japan","Russia","Germany","Australia"],
    'Days':["sunday","saturday","tuesday","thursday","monday","friday","sunday","friday","tuesday","monday"],
    'Speed Limit':[37,96,62,78,98,30,92,61,106,74]
}
accident_list=pd.DataFrame(data)
accident_list.head()
```

```python
input_data=accident_list[['Days','country']]
target_data=accident_list['Speed Limit']
```

```python
model=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,target_data,test_size=0.2,random_state=32)
```

```python
# Apply one-hot encoding to the 'Days' and 'country' columns in input_data
input_data_encoded = pd.get_dummies(input_data, columns=['Days', 'country'], drop_first=True)

# Fit the model with the encoded input data
model.fit(input_data_encoded,target_data)
```

```python
# Get the feature names the model was trained on from input_data_encoded
trained_features = input_data_encoded.columns

# One-hot encode x_test for prediction, using the same columns as were used for input_data_encoded
x_test_encoded = pd.get_dummies(x_test, columns=['Days', 'country'], drop_first=True)

# Align columns to match the training data's structure and order
# This handles cases where some categories might be present in trained_features but not in x_test_encoded (or vice versa)
missing_cols_in_test = set(trained_features) - set(x_test_encoded.columns)
for c in missing_cols_in_test:
    x_test_encoded[c] = 0

# Ensure the order of columns in x_test_encoded is identical to trained_features
x_test_encoded = x_test_encoded[trained_features]

predicted_output=model.predict(x_test_encoded)
predicted_output
```

```python
mse=mean_squared_error(y_test,predicted_output)
print("Mean Squared Error:",mse)
r2=r2_score(y_test,predicted_output)
print("R2 Score:",r2)
```
